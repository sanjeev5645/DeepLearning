```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import cv2
import PIL.Image as Image
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/DATA/car1.webp")
cv2_imshow(image)
```

```python

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
pretrained = '/content/DATA/haarcascade_russian_plate_number.xml'
model = cv2.CascadeClassifier(pretrained)

```

```python
model1 = model.detectMultiScale(gray_image, scaleFactor = 1.05, minNeighbors = 5, minSize = (40,40))
```

```python
print(model1)
```
